## Deep Learning: TensorFlow

```python
import tensorflow as tf
print("TensorFlow version:", tf.__version__)
```

```python
# Ejemplo
# TensorFlow V2
a = tf.constant(1.5)
b = tf.constant(2.5)

c = a + b 
tf.print(c)
print(c)
```

```python
# Ejemplo 
# TensorFlow V1
tf.compat.v1.disable_eager_execution() # Desactivar la ejecución para usar el modo grafo

d = tf.compat.v1.placeholder(tf.float32)
e = tf.compat.v1.placeholder(tf.float32)

add = tf.add(d, e)

sess = tf.compat.v1.Session() # Iniciar la sesión que ejecuta el grafo


binding = {d: 1.5, e: 2.5}

f = sess.run(add, feed_dict=binding) # Ejecutar la operación de suma
print(f)
sess.close()
```
